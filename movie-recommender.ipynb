```python
def create_tags(obj):
    li=[]
    for i in ast.literal_eval(obj):
        li.append(i['name'])
    return li

        
```

```python
movies['genres'].apply(create_tags)#.apply act as map function in pandas dataframe ,
```

```python
movies['genres']=movies['genres'].apply(create_tags)
```

```python
movies.head(1)
```

```python
movies['keywords']=movies['keywords'].apply(create_tags)
```

```python
movies['cast'][0]
```

```python
def create_tags2(obj): #these function takes name of only 3 actors
    li=[]
    count=0
   
    for i in ast.literal_eval(obj):
        while count<=3:
            li.append(i['name'])
            count+=1
            break
        
        
    return li
```

```python
movies['cast']=movies['cast'].apply(create_tags2)
```

```python
movies.head()
```

```python
movies['crew'][0]
```

```python
def fetch_director(obj): #these function takes name of director
    li=[]
   
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            li.append(i['name'])
            break
        
    return li
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

### Removing spaces between genres ,crew and cast (to avoid conflict in names)

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","")for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","")for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movies
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
new_df=movies[['movie_id','title','tags']]
```

```python
 new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

# Text Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names()
```

```python
# action and actions appears two features ,similarly actor,actors
#using stemming to remove these
#example: [ride,ridden,rode]
#         [ride,ride,ride]
```

```python
#NLP library NLTK
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
# again doing text vectorization after stemming(we don't have to do this always,direct stemming and text vectorization)
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.get_feature_names()
```

```python
#every movie is a vector now.
#we need to calculate distance with movies to get the similiraties
#5000 dimensions (features) so euclidean distance is uneffective
#we use cosine distance which measures the angle between them.

 
```

## cosine similarity

```python
#Cosine similarity says that to find the similarity between two points or vectors we need to find Angle between them.
# it is measured between 0-1
#when points are near higher value,when points are farther lower value
'''When angle between points P1 & P2 is 45 Degree then
cosine_similarity= Cos 45 = 0.525

Case 2: When two points P1 & P2 are far from each other and angle between points is 90 Degree then

cosine_similarity= Cos 90 = 0
Case 3: When two points P1 & P2 are very near and lies on same axis to each other and angle between points is 0 Degree then

cosine_similarity= Cos 0 = 1

'''
from sklearn.metrics.pairwise import  cosine_similarity  
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity
```

```python
similarity.shape #it is relating similiraty to all movies,thats why 4806*4806
```

```python
# (similarity[0]+similarity[1])
```

```python
def recommend(movie):
    movie_index=new_df[new_df['title']==movie].index[0]
    distances=similarity[movie_index]
    movie_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:50]
    
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
        
        
        
        
        
    
```

```python
def recommend2(*movie):
    movie_indices=[]
    for i in movie:
        movie_indices.append(new_df[new_df['title']==i].index[0])
    print(len(movie))
    sum=0
    #for j in range(0,len(movie)):
        #sum=sum+similarity[movie_indices[j]]
    #distances=sum/len(movie)
    for j in movie_indices:
        sum+=similarity[j]
    distances=sum/len(movie)
    
    
    
    #distances=(similarity[movie_index1]+similarity[movie_index2])/2
    movie_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:50]
   # print(movie_list)
    
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```
